```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv")
    #index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(df_market_data[['price_change_percentage_24h', 'price_change_percentage_7d',	'price_change_percentage_14d',	'price_change_percentage_30d',	'price_change_percentage_60d',	'price_change_percentage_200d',	'price_change_percentage_1y']])
```

```python
# Create a DataFrame with the scaled data
market_data_scaled_df = pd.DataFrame(market_data_scaled, 
                                     columns=['price_change_percentage_24h', 'price_change_percentage_7d',
                                              'price_change_percentage_14d', 'price_change_percentage_30d',
                                              'price_change_percentage_60d', 'price_change_percentage_200d', 
                                              'price_change_percentage_1y'])

# Copy the crypto names from the original DataFrame
market_data_scaled_df['coin_id'] = df_market_data['coin_id'].values  # Use .values to ensure it's a plain array

# Set the coin_id column as index
market_data_scaled_df = market_data_scaled_df.set_index('coin_id')

# Display the scaled DataFrame
market_data_scaled_df.head()
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
# Create a a list to store the values of k

k = list(range(1, 11))

```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k:

  k_model = KMeans(n_clusters=i, random_state=0)

  k_model.fit(market_data_scaled_df)

  inertia.append(k_model.inertia_)

  
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k":k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Review the Dataframe
df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x='k',
    y='inertia',
    title='Elbow Curve',
    xticks=k
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 4

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
model_4 = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled DataFrame
model_4.fit(market_data_scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
k_higher = model_4.predict(market_data_scaled_df)

# Print the resulting array of cluster values.
print(k_higher)
```

```python
# Create a copy of the scaled DataFrame
market_data_scaled_copy = market_data_scaled_df.copy()
```

```python
# Check the columns in the original DataFrame
print(market_data_scaled_df.columns)
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
market_data_scaled_copy['clusters_higher'] = k_higher

# Display the copy of the scaled DataFrame
market_data_scaled_copy.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
# Create the scatter plot
scatter_plot = market_data_scaled_copy.hvplot.scatter(
    x="price_change_percentage_24h",  # x-axis: 24h price change percentage
    y="price_change_percentage_7d",   # y-axis: 7d price change percentage
    by="clusters_higher",             # Color points by cluster label
    hover_cols=["coin_id"],       # Display cryptocurrency name on hover
).opts(
    yformatter="%.5f",  # Format the y-axis labels to be integers (no decimals)
    title="Price Change Percentages vs. Clusters",  # Set the title of the plot
    xlabel="Price Change (24h)",  # Label for x-axis
    ylabel="Price Change (7d)",   # Label for y-axis
    height=600,  
    width=800    
)

# Display the plot
scatter_plot
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3) # how many principle components it will have
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
# Fit the PCA model on the transformed market DataFrame

market_pca = pca.fit_transform(market_data_scaled_df)

# View the scaled PCA data
market_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 0.89503166

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create a new DataFrame with the PCA data
# Create a new DataFrame with the PCA data and set the index to 'coin_id'
market_pca_df = pd.DataFrame(
    market_pca,  # PCA results
    columns=["PCA1", "PCA2", "PCA3"],  # Column names for the PCA components
    index=market_data_scaled_df.index  # Set the original index ('coin_id') from the scaled DataFrame
)

# Display the scaled PCA DataFrame 
market_pca_df.head()
```

```python
# Create an empty list to store the inertia values
inertia_values_PCA = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(market_pca_df[['PCA1', 'PCA2', 'PCA3']])  # Use all 3 components
    inertia_values_PCA.append(kmeans.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_PCA = {'k': k_values, 'inertia': inertia_values}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow_PCA = pd.DataFrame(elbow_data_PCA)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
df_elbow_PCA.hvplot.line(
  x="k",
  y="inertia",
  title="Elbow Curve",
  xticks=k

)
```

  * **Answer:** 4

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
k_4_PCA = model.predict(market_pca_df)

# Print the resulting array of cluster values.
k_4_PCA[:5]
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
market_pca_predictions_df.hvplot.scatter(
  x="PCA1",  # Principal Component 1
  y="PCA2",  # Principal Component 2
  by="clusters_higher_PCA",  # Color points by the clusters found using K-Means
  hover_cols=["coin_id"],  # Add the cryptocurrency name in the hover column
  title="Scatter Plot of Clusters (PCA Data)",  # Title of the plot
  xlabel="Principal Component 1",  # Label for x-axis
  ylabel="Principal Component 2"  # Label for y-axis
)

```
